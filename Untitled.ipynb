```python
df.head()
```

```python
df.drop('Loan_ID', axis=1, inplace=True)
```

```python
# Correlation
sns.heatmap(df.corr(),annot=True)
```

```python
X=df.drop(labels=['Loan_Status'],axis=1)
y = df[['Loan_Status']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# i.e. 70 % training dataset and 30 % test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train.head()
```

```python
# creating a RF classifier
random_forest_c= RandomForestClassifier(n_estimators = 100)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
random_forest_c.fit(X_train, y_train)
  
# performing predictions on the test dataset
y_pred = random_forest_c.predict(X_test)
  
# metrics are used to find accuracy or error
from sklearn import metrics  
print()
  
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
logestic_regression=LogisticRegression(random_state=None)
logestic_regression.fit(X_train,y_train)
y_pred=logestic_regression.predict(X_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
design_tree=DecisionTreeClassifier()
design_tree.fit(X_train,y_train)
y_pred=design_tree.predict(X_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```
